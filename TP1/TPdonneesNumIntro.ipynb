#### exponentielle

Pour le premier dessin, tracez la courbe qui représente la fonction exponentielle pour les valeurs entières entre 0 et 500.

```python
# dessin 1
x = np.arange(0, 500)
y = np.exp(x)
plt.plot(x, y)
```

**Peut-on aller vers de plus grandes valeurs ? Quelle est la limite ? Pourquoi ?**

Oui, nous pouvons aller vers de plus grandes valeurs, la limite étant le nombre maximum que permettent de représenter les réels (float) en python. Ces float sont codés sur 64 bits et les nombres représentables sont représentés en se basant sur la norme IEEE754. La limite en termes d'exposant est donc 10^308 dans les positifs. Il s'agit donc de la limite des nombres que l'on peut représenter.

Ajoutez des légendes sur les axes de la figure.

```python
# dessin 1
x = np.arange(0, 500)
y = np.exp(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphique représentant les nombres de 0 à 499 à l\'exponentielle')
plt.plot(x, y)
```

```python
# dessin 2
# tracez la même courbe mais pour des valeurs entières entre 0 et 10
x = np.arange(0, 10)
y = np.exp(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphique représentant les nombres de 0 à 10 à l\'exponentielle')
plt.plot(x, y)
```

**Les deux graphiques précédents n'ont pas la même allure. Pourquoi ?**

Les deux graphiques n'ont pas la même échelle.
Le premier est à l'échelle 10^216, alors que le deuxième est à l'échelle 10^4.
Ainsi, puisque sur le premier graphique, l'échelle est de 10^216, tout ce qui est en dessous de 1 * 10^216 sera à peu près au même niveau, c'est-à-dire proche de 0. C'est le cas des nombres de 0 à 495 environ (sûrement 497).
Alors que dans l'autre graphique, l'échelle étant de 10^4

Pour le dessin numéro 2, remplacer les valeurs entières de x par des valeurs décimales de telle sorte qu'il y ait une centaine de valeurs entre 0 et 10.

```python
x = np.linspace(0, 10, num=500)
y = np.exp(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphique représentant les nombres de 0 à 10 à l\'exponentielle avec une précision accrue')
plt.plot(x, y)
```

Pour chaque valeur entière et impaire de x, tracez un petit segment qui passe par ce point (xi,yi) et de pente exp(xi). A quoi cela correspond ?

```python
# à vous
x = np.linspace(0, 10, num=500)
y = np.exp(x)
for i in range(1, int(max(x)), 2):
    x1 = i - 1 / 2
    x2 = i
    x3 = i + 1 / 2
    point1 = [x1, np.exp(i) * (x1 - i) + np.exp(i)]
    point2 = [x2, np.exp(i) * (x2 - i) + np.exp(i)]
    point3 = [x3, np.exp(i) * (x3 - i) + np.exp(i)]
    x_values = [point1[0], point2[0], point3[0]]
    y_values = [point1[1], point2[1], point3[1]]
    plt.plot(x_values, y_values, 'o', linestyle="-")
plt.xlabel('x')
plt.ylabel('y')
plt.title("exp(x)")
plt.plot(x, y)
```

Ces segments représentent les tangentes à la courbe au niveau des valeurs étudiées.

On sait qu'un segment est représentable par une fonction affine, ainsi une telle équation de segment peut s'écrire : y = ax+b, avec a
qui représente le coefficient directeur (la pente), et b l'ordonnée à l'origine.
Ici, on sait déjà que a = exp(xi).
Pour b, cela demande un peu de travail pour le trouver.
A ce moment du raisonnement, on a : y = exp(xi) * x + b.
Autrement dit, b = y - exp(xi) * x.
Il faut ici remplacer x et y par les valeurs d'un point appartenant au segment qu'on souhaite créer.
Un point qui appartient à chaque segment que l'on veut créer est le point de coordonnées (xi, yi).
Ainsi, on obtient : b = yi - exp(xi) * xi.
Si on revient sur notre équation affine originale, on a alors désormais : y = exp(xi) * x + (yi - exp(xi) * xi).
On peut simplifier cela en factorisant un peu, on obtient alors : y = exp(xi)(x-xi) + yi.
Cette équation de droite/segment est alors celle qui va fonctionner pour chaque point de notre courbe pour lequel on veut tracer une droite passant par le point (xi, yi) avec une pente de exp(xi).

Cette équation affine peut également s'écrire : f(x) = exp(xi)(x - xi) + yi, ou, en développant encore un peu : f(x) = exp(xi)(x - xi) + exp(xi) = exp(xi)(x - xi + 1).

J'utilise ensuite simplement cette propriété afin de tracer des segments passant par trois points ayant pour abscisse xi - 1/2, xi et xi + 1/2, afin d'avoir un segment de taille 1 passant par le point de coordonnées (xi, yi).

#### logarithme

Explorez de la même façon la fonction logarithme. Jouez avec les valeurs de x. Est-ce un logarithme népérien ou en base 10?

```python
# à vous
x = np.linspace(1, 500, num=100)
y = np.log(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphique représentant les logarithmes des nombres de 1 à 500')
plt.plot(x, y)
```

Ajoutez les segments de pente 1/xi:

```python
# à vous
x = np.arange(0.1, 10, 0.1)
y = np.log(x)
for i in range(1, int(max(x)) + 1, 2):  # +1 afin de tomber sur 10
    x1 = i - 1 / 2
    x2 = i
    x3 = i + 1 / 2
    point1 = [x1, (x1 / i) + np.log(i) - 1]
    point2 = [x2, (x2 / i) + np.log(i) - 1]
    point3 = [x3, (x3 / i) + np.log(i) - 1]
    x_values = [point1[0], point2[0], point3[0]]
    y_values = [point1[1], point2[1], point3[1]]
    plt.plot(x_values, y_values, 'o', linestyle="-")
plt.xlabel('x')
plt.ylabel('y')
plt.title("log(x)")
plt.plot(x, y)
```

Explications du raisonnement de réalisation :

Même raisonnement que pour l'exponentielle, voici les calculs :

y = ax + b, avec a = 1/xi ici.
donc y = (1/xi * x) + b = x/xi + b.
Donc b = y - x/xi.
On prend un point qu'on veut faire appartenir au segment : le point de coordonnées (xi, yi), et on applique ses coordonnées à x et y, ce qui nous donne : b = yi - xi/xi = yi - 1.
Revenons à notre équation affine originale : y = x/xi + b = x/xi + (yi - 1).
On a donc notre fonction affine/segment qui est representé par l'équation suivante :

y = x/xi + yi - 1.
Ici, yi = log(xi), nous obtenons donc finalement : y = x/xi + log(xi) - 1.

Je crée ensuite chaque segment au niveau de chaque valeur entière impaire en faisant passer chaque segment par trois points, dont notre point de coordonnées (xi, yi).

#### combiner log et exp

```python
# tracer la fonction f(x) = log(exp(x))
x = np.linspace(0, 500, num=100)
y = np.log(np.exp(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphique représentant les logarithmes des nombres de 0 à 500 à l\'exponentielle')
plt.plot(x, y)
```

On remarque alors que nous obtenons une fonction linéaire y = x. Cela signifie que np.log fait bien appel à un logarithme népérien, en base e, qui est la réciproque de l'exponentielle.

#### sinusoïde

Mêmes questions.

```python
# à vous
x = np.linspace(0, 500, num=100)
y = np.sin(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphique représentant les sinus des nombres de 0 à 500')
plt.plot(x, y)
```

```python
# avec les segments de pente ...
x = np.arange(0.1, 10, 0.1)
y = np.sin(x)
for i in range(1, int(max(x)) + 1, 2):  # +1 pour atteindre 10
    x1 = i - 1 / 2
    x2 = i
    x3 = i + 1 / 2
    point1 = [x1, np.cos(i) * (x1 - i) + np.sin(i)]
    point2 = [x2, np.cos(i) * (x2 - i) + np.sin(i)]
    point3 = [x3, np.cos(i) * (x3 - i) + np.sin(i)]
    x_values = [point1[0], point2[0], point3[0]]
    y_values = [point1[1], point2[1], point3[1]]
    plt.plot(x_values, y_values, 'o', linestyle="-")
plt.xlabel('x')
plt.ylabel('y')
plt.title("sin(x)")
plt.plot(x, y)
```

Explications sur le raisonnement :

Je vais simplement mettre les équations, le principe est exactement le même que précédemment...
La pente est la derivée de la fonction étudiée. La dérivée de sin(x) est cos(x). Appliqué à xi comme précédemment...on obtient cos(xi).
y = ax + b, où a = cos(xi).
y = cos(xi)*x + b => b = y - cos(xi) * x.
On remplace x et y par xi et yi, qui sont les coordonnées d'un point appartenant au segment qu'on veut créer.
b = yi - cos(xi) * xi = sin(xi) - cos(xi) * xi.

Revenons à la fonction affine initiale : y = cos(xi) * x + b = cos(xi) * x + (sin(xi) - (cos(xi) * xi)).
Cela se simplfie par une factorisation en cos(xi) : y = cos(xi)(x - xi) + sin(xi).

On trace ensuite chaque segment en le faisant passer par les points de coordonnées (xi, yi).

```python
#chargement des librairies
import librosa
import librosa.display
import IPython.display as ipd
```

Pour un premier son, générez un 'la' de référence 440Hz.

```python
#un petit test de son
n = 6400  # nombre d'échantillons
te = 1 / 8000.0  # période d'échantillonnage
fe = 8000  # fréquence d'échantillonnage
t = np.linspace(0, n * te, n, endpoint=False)
# accordons-nous sur un la 440 Hz
s = np.cos(2 * np.pi * 440 * t)
```

Combien de temps dure ce son?

Ce son dure légérement moins d'une seconde, il s'agit du "bip" d'un téléphone lorsqu'on est dans l'attente de joindre
quelqu'un lors d'un appel.
En effet, il y a 6400 échantillons, et la fréquence d'échantillonnage est de 8000 Hz. Nous allons donc lire 8000 échantillons en 1 seconde.
Puis nous avons 6400 échantillons, nous aurons fini de lire ces échantillons avant 1 seconde (6400/8000 = 0.8, donc nous aurons fini de
lire les 6400 échantillons en 0.8 secondes.

```python
# écoutons 
ipd.Audio(s, rate=fe, autoplay=True)
```

Testez d'autres fréquences et d'autres combinaison de sinusoïdes. Par exemple la somme d'un son pur à 440 Hz et d'un autre à 200 Hz de puissance double.

```python
s = 0.25 * np.cos(2 * np.pi * 440 * t + 0.5) + 0.5 * np.cos(2 * np.pi * 200 * t)
```

Comment faire durer le son plus longtemps? Par exemple 4 secondes.

```python
ipd.Audio(s, rate=0.2 * fe, autoplay=True)
```

Pour faire durer le son plus longtemps, il est possible de réduire la fréquence d'échantillonnage du son. En multipliant fe par 0.2, on donne finalement une fréquence d'échantillonage de 1600 Hz. Or nous avons 6400 échantillons, le programme a alors besoin de 4 secondes pour lire les 6400 échantillons à cette fréquence (1600 échantillons lus par seconde * 4 secondes = 6400 échantillons lus au bout de 4 secondes).

## 3. Quantification et échantillonnage
