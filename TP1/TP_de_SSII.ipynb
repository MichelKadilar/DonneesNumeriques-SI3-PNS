### 1.2 tracés de fonctions mathématiques usuelles

#### exponentielle

Pour le premier dessin, tracez la courbe qui représente la fonction exponentielle pour les valeurs entières entre 0 et 500.

```python
# dessin 1
x = np.arange(0,500)
y = np.exp(x)
plt.plot(x,y)
```

Peut-on aller vers de plus grandes valeurs? Quelle est la limite? Pourquoi?

Car la valeur ordonnée vaut l'exponentielle de la valeur en X. Etant donné qu'exp(500) représente une très très grande valeur et que la fonction exponentielle croît très vite, il est normal d'obtenir une courbe aussi droite. La limite doit correspondre a la limite d'encodage des nombres sur Python.

Ajoutez des légendes sur les axes de la figure.

```python
# dessin 1
x = np.arange(0,500)
y = np.exp(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('exp(x)')
plt.plot(x,y)
```

```python
# dessin 2
# tracez la même courbe mais pour des valeurs entières entre 0 et 10
x = np.arange(0, 10)
y = np.exp(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title("exp(x)")
plt.plot(x, y)
```

Les deux graphiques précédents n'ont pas la même allure. Pourquoi?
Pour le dessin numéro 2, remplacer les valeurs entières de x par des valeurs décimales de telle sorte qu'il y ait une centaine de valeurs entre 0 et 10.

Les graphiques n'ont pas la même allure car les valeurs prise pour ce dessin vont uniquement de 0 à 10, ce qui est bien moins grand que de 0 à 500. Ainsi, on peut encore apercevoir la courbe de la fonction exponentielle.

```python
# Ils n'ont pas la même allure car les valeurs du premier graphique sont extrêmement élevées à mesure que x augmente. Elles s'approchent de l'infini
# et donc rendent la courbe (ou pente) plus aiguë à mesure que l'ont dézoom.
# Le second graphique possède une pente moins "agressive" étant donné que les valeurs de f(x) = y sont plus faibles. 

x = np.arange(0, 10, 0.1)
y = np.exp(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title("exp(x)")
plt.plot(x, y)
                
```

Pour chaque valeure entière et impaire de x, tracez un petit segment qui passe par ce point (xi,yi) et de pente exp(xi). A quoi cela correspond?

```python
x = np.arange(0, 10, 0.1)
y = np.exp(x)
plt.plot(x,y)
for i in range (1,10,2):
    x = [i-1/2, i+1/2]
    y = [1/2 * np.exp(i), np.exp(i) + 1/2 * np.exp(i)]
    plt.plot(x,y,color="red")
    plt.scatter(x,y,color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.title("exp(x)")
plt.plot(x, y)
```

#### logarithme

Explorez de la même façon la fonction logarithme. Jouez avec les valeurs de x. Est-ce un logarithme népérien ou en base 10?

C'est un log népérien car exp(log(x)) donne une constante

```python
# Ils n'ont pas la même allure car les valeurs du premier graphique sont extrêmement élevées à mesure que x augmente. Elles s'approchent de l'infini
# et donc rendent la courbe (ou pente) plus aiguë à mesure que l'ont dézoom.
# Le second graphique possède une pente moins "agressive" étant donné que les valeurs de f(x) = y sont plus faibles. 

x = np.arange(0, 10, 0.1)
y = np.log(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title("log(x)")
plt.plot(x, y)
```

Ajoutez les segments de pente 1/xi:

```python
x = np.arange(0.1, 10, 0.1)
y = np.log(x)
plt.plot(x,y)
for i in range (1,10,2):
    x = [i-1/2, i+1/2]
    y = [np.log(i) - 1/2 * 1/i, np.log(i) + 1/i * 1/2]
    plt.plot(x,y,color="red")
    plt.scatter(x,y,color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.title("log(x)")
plt.plot(x, y)
```

#### combiner log et exp

```python
# tracer la fonction f(x) = log(exp(x))
x = np.arange(0.1, 10, 0.1)
y = np.log(np.exp(x))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("log(exp)(x))")
```

```python
#chargement des librairies
import librosa
import librosa.display
import IPython.display as ipd
```

Pour un premier son, générez un 'la' de référence 440Hz.

```python
#un petit test de son
n=6400# nombre d'échantillons
te=1/8000.0 # période d'échantillonnage
fe=8000 # fréquence d'échantillonnage
t = np.linspace(0, n*te, n,endpoint=False)
# accordons-nous sur un la 440 Hz
s= np.cos(2*np.pi*440*t)
```

Combien de temps dure ce son?

fe/n = 1,25 s

```python
# écoutons 
ipd.Audio(s,rate=fe,autoplay=True)
```

Testez d'autres fréquences et d'autres combinaison de sinusoïdes. Par exemple la somme d'un son pur à 440 Hz et d'un autre à 200 Hz de puissance double.

```python
s = 0.25*np.cos(2*np.pi*440*t+0.5) + 0.5*np.cos(2*np.pi*200*t)
```

```python
# regardons
# 2 types de visualisation: par librosa.display.waveplot et par plt.plot
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,5))
librosa.display.waveshow(s,sr=fe, ax=ax1)
ax2.set_xlabel('temps (s)')
ax2.set_ylabel('amplitude')
ax2.set_title("chronogramme")
ax2.plot(t,s)
plt.show(s[50])
plt.title("chronogramme")
plt.show()
```

Voyez-vous les belles sinusoïdes? Pourquoi? Comment modifier la visualisation? 

On voit des sinusoïdes assez jolies. On peut modifier la visualisation en modifiant les paramètres de subplot.

```python
# Le nombre d'échantillons est trop grand pour que l'on puisse voir les sinusoïdes

n=100
te=1/8000.0
fe=8000 
t = np.linspace(0, n*te, n,endpoint=False)
s= np.cos(2*np.pi*440*t)
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20, 5))
librosa.display.waveshow(s,sr=fe, ax=ax1)
ax2.set_xlabel('temps (s)')
ax2.set_ylabel('amplitude')
ax2.set_title("chronogramme")
ax2.plot(t,s)
plt.title("chronogramme")
plt.show()
```
